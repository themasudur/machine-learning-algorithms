###জিনি/গিনি ইন্ডেক্স/ইম্পিউরিটি

**ফরমুলা**

$1-(p^2+q^2)$

এখানে $p$, $q$ আলাদা ক্লাস বুঝাচ্ছে। কোন নির্দিষ্ট এট্রিবিউটের জন্য প্রতিটি ক্লাসের করেস্পন্ডিং স্যাম্পল সংখ্যার হিসেবে $p$, $q$ বের করা হয়। 

**উদাহরণ**

*প্লে টেনিস* ডেটা সেটের জন্য, Outlook এট্রিবিউটের জিনি ইম্পিউরিটি বের করি- 

ফিচার | ইয়েস | নো | টোটাল | জিনি ইম্পিউরিটির হিসেব | 
--- | --- | --- | --- | --- |
Sunny | 2 | 3 | 5 | $1- (\frac{2}{5})^2 - (\frac{3}{5})^2 = 0.48$ |
Overcast | 4 | 0 | 4 | $1- (\frac{4}{4})^2 - (\frac{0}{4})^2 = 0$ |
Rainy | 3 | 2 | 5 | $1- (\frac{3}{5})^2 - (\frac{2}{5})^2 = 0.48$ |

এখন, Outlook এট্রিবিউটের জিনি ইম্পিউরিটি = $\frac{5}{14}*gini(Sunny) + \frac{4}{14}*gini(Overcast) + \frac{5}{14}*gini(Rainy)$
$= \frac{5}{14}*0.48 + \frac{5}{14}*0 + \frac{5}{14}*0.48 = 0.3429$

এভাবে অন্য এট্রিবিউটগুলোর জন্যও জিনি ইম্পিউরিটি বের করবো। সেখান থেকে সবচেয়ে ছোটটা নিবো। রিকার্সিভলি এই কাজ চলতে থাকবে। 

>(কাজ ৯.১ বাকি এট্রিবুটগুলোর জন্য *জিনি/গিনি ইন্ডেক্স/ইম্পিউরিটি* নিজেরা বের করুন)

**এলগোরিদম** 

জিনি/গিনি ইন্ডেক্স/ইম্পিউরিটি এর একটি জনপ্রিয় এলগোরিদম হলো-

###কাই-স্কয়ার

এটি মূলত সাবট্রি আর এর রুটের মধ্যে স্ট্যাস্টিক্যাল পার্থক্য বের করে। 

**ফরমুলা**

$\sqrt{\frac{(Actual - Expected)^2}{Expected}}$

Actual হলো টোটাল একচুয়াল স্যাম্পল সংখ্যা এবং Expected হলো Actual/(ক্লাস সংখ্যা)। 

**উদাহরণ**

প্লে টেনিসে ডেটা সেটের জন্য, Outlook এট্রিবিউটের জিনি ইম্পিউরিটি বের করি- 

ফিচার | ইয়েস | নো | টোটাল | এক্সপেক্টেড | কাই স্কয়ারের হিসেব | 
--- | --- | --- | --- | --- | --- |
Sunny | 2 | 3 | 5 | 2.5 |  $\sqrt{\frac{(5 - 2.5)^2}{2.5}} = 1.581$ |
Overcast | 4 | 0 | 4 | 2 | $\sqrt{\frac{(4 - 2)^2}{2}} = 1.41$ |
Rain | 3 | 2 | 5 | 2.5 | $\sqrt{\frac{(5 - 2.5)^2}{2.5}} = 1.581$ |

এদের যোগফল = 0.316 + 1.41 + 0.316 = 2.042

বাকি এট্রিবিউট গুলোর জন্য করে সবচে বড়টা নিবো। 

**এলগোরিদম** 

কাই স্কয়ার এর একটি জনপ্রিয় এলগোরিদম হলো-  
CHAID (Chi-square Automatic Interaction Detector)

**এর কিছু বৈশিষ্ট্য:**
* শুধুমাত্র স্ট্যাস্টাটিক 
* মাল্টি ক্লাসিফিকেশান ভালো পারফরম্যান্স

###ইনফরমেশান গেইন

**ফরমুলা**

$-(p_+ \log_2 p_+) + -(p_- \log_2 p_-)$

এখানে $p_+$, $p_-$ আলাদা ক্লাস বুঝাচ্ছে। কোন নির্দিষ্ট এট্রিবিউটের জন্য প্রতিটি ক্লাসের করেস্পন্ডিং স্যাম্পল সংখ্যার হিসেবে $p_+$, $p_-$ বের করা হয়। 

**উদাহরণ**

প্লে টেনিসে ডেটা সেটের জন্য, Outlook এট্রিবিউটের জিনি ইম্পিউরিটি বের করি- 

ক্লাস লেবেলের  জন্য $gain(PlayTennis) = -\frac{9}{14}*\log_2{\frac{9}{14} + -\frac{5}{14}*\log_2{\frac{5}{14}}} = 0.94$ 

ফিচার | ইয়েস | নো | টোটাল | ইনফরমেশান গেইন এর হিসেব | 
--- | --- | --- | --- | --- |
Sunny | 2 | 3 | 5 | $-\frac{2}{5}*\log_2{\frac{2}{5}} + -\frac{3}{5}*\log_2{\frac{3}{5}} = 0.97$ |
Overcast | 4 | 0 | 4 | $-\frac{4}{4}*\log_2{\frac{4}{4}} + -\frac{0}{4}*\log_2{\frac{0}{4}} = 0$ |
Rainy | 3 | 2 | 5 | $-\frac{3}{5}*\log_2{\frac{3}{5}} + -\frac{2}{5}*\log_2{\frac{2}{5}} = 0.97$ |

এখন, Outlook এট্রিবিউটের ইনফরমেশান গেইন = $gain(PlayTennis) - \{\frac{5}{14}*gain(Sunny) + \frac{4}{14}*gain(Overcast) + \frac{5}{14}*gain(Rainy)$
$= \frac{5}{14}*0.97 + \frac{4}{14}*0 + \frac{5}{14}*0.97 = 0.247$

এভাবে অন্য এট্রিবিউটগুলোর জন্যও ইনফরমেশান গেইন বের করবো। সেখান থেকে সবচেয়ে বড়টা নিবো। রিকার্সিভলি এই কাজ চলতে থাকবে। 

> (কাজ ৯.২ বাকি এট্রিবুটগুলোর জন্য *ইনফরমেশান গেইন* নিজেরা বের করুন)

##কোড পরে

```
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn import tree 
from sklearn.preprocessing import LabelEncoder

import pandas as pd 
import numpy as np 

df = pd.read_csv('https://raw.githubusercontent.com/m4jidRafiei/Decision-Tree-Python-/master/playtennis.csv')

lb = LabelEncoder() 
df['outlook_'] = lb.fit_transform(df['Outlook']) 
df['temp_'] = lb.fit_transform(df['Temperature'] ) 
df['humidity_'] = lb.fit_transform(df['Humidity'] ) 
df['windy_'] = lb.fit_transform(df['Wind'] )   
df['play_'] = lb.fit_transform(df['Play Tennis'] ) 
X = df.iloc[:,5:9] 
Y = df.iloc[:,9]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100) 

clf_entropy = DecisionTreeClassifier(criterion='gini')
clf_entropy.fit(X_train.astype(int),y_train.astype(int)) 
y_pred_en = clf_entropy.predict(X_test)

print("Accuracy is: {}".format(accuracy_score(y_test.astype(int),y_pred_en) * 100))
```

### বেইজ থিউরেম

কন্ডিশনাল প্রোবাবিলিটি এর উপর কাজ করে যেখানে প্রিওর আর পোস্টেরিওর প্রোবাবিলিটির কথা আছে। 

একটি ইভেন্টের প্রোবাবিলিটি দেওয়া থাকলে আরেকটি ইভেন্টের প্রোবাবিলিটি বের করার কাজকেই কন্ডিশনাল প্রোবাবিলিটি বলে। 

P(A given B) or P(A | B)

কন্ডিশনাল প্রোবাবিলিটি এর ফরমুলা
$P(A|B) = \frac{P(B|A) * P(A)}{P(B)}$

এখানে, 
* P(A|B): পোস্টেরিওর প্রোবাবিলিটি
* P(A): প্রিওর প্রোবাবিলিটি
* P(B|A): লাইকিহুড
* P(B): এভিডেন্স

তাহলে ফরমুলা দাঁড়ালো, 

পোস্টেরিওর= {লাইকিহুড * প্রিওর}/{এভিডেন্স}

**P(Yes | Overcast) বের করি** 

P(Yes | Overcast) = { P(Overcast | Yes) * P(Yes) }/ P (Overcast)

এখানে,

P(Overcast) = 4/14 

P(Yes)= 9/14

P(Overcast | Yes) = 4/9

তাহলে,

```
# গাউসিয়ান
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

y_pred = GaussianNB().fit(X_train, y_train).predict(X_test)

print("Number of misclassified points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
# Task 9.8- which are mislabeled? Find them.
```
