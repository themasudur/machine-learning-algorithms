**2.1 Dealing with Missing Values**

Missing data = unclean data. These values usually take the form of NaN or None.

Several ways of handling missing values:

* Remove the instances *if* you have big enough data set and the percentage of missing values is high (over 50%)
* Dropping attributes with missing values
* Imputing the attribute (mean/ median/ mode): Use the Imputerclass from the `scikit-learn` library
* Imputing the missing values via linear regression

#Step 3: Preparing the Features for Machine Learning

**3.1 Handling categorical data**

ML deals with only numeric values (float or int data type). However, data sets often contain categorical data that needs to be transformed into numeric. 

Categorical values are discrete and can be encoded as dummy variables, assigning a number for each category. `Scikit-learn` library automatically converts categorical values to nummeric array. 

The simplest way is to use Label Encoder/One Hot Encoder, specifying the index of the column.

```
import pandas as pd
play = pd.read_csv('play_tennis.csv')
play
```

```
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
play['outlook'] = labelencoder.fit_transform(play['outlook'])
play
```

**3.2 Seperating Target/Class column**

Divide the data set into *train* and *test* set.  



That first colon `:` means that taking all the records in our dataset. 

`:-1` means that taking all of the columns except the last one.

`-1` means the last column only.

`.values` means considering all values.


```
print(play)
```

```
X = play.iloc[:, :-1].values 
print(X)
```

```
y = play.iloc[:, -1].values
print(y)
```

**3.3 Spliting dataset into train and test sets**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

To ensure same subset always: use `random_state`

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train
```

**Scaling and Normalization**

Standardization

Usually a dataset contains variables that are different in scale. 

e.g. an Employee dataset will contain AGE column with values on scale 20-70 and SALARY column with values on scale 10000-80000.


As these two columns are different in scale, they are Standardized to have common scale while building machine learning model.

**StandardScaler:** transform your data such that its distribution will have a mean value 0 and standard deviation of 1.

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

calhouse = pd.read_csv('sample_data/california_housing_train.csv')

X = calhouse.iloc[:, :-1].values 
y = calhouse.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape)
print(X_train)

# Scaling
scale = StandardScaler()
X_train = scale.fit_transform(X_train)

print(X_train.shape)
print('Train data: ', X_train)
```
